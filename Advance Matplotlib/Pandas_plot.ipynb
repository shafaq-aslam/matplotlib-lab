# Pandas Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# on a Series
series = pd.Series([1, 8, 3, 4, 2, 2])
series.plot(kind = 'line')

# can be used on a dataframe as well
```

```python
tip = sns.load_dataset('tips')
tip.head()
```

### Scatter plot

```python
tip.plot(kind = 'scatter', x = 'total_bill', y = 'tip')
```

```python
# title
tip.plot(kind = 'scatter', x = 'total_bill', y = 'tip', title = 'Restaurant Bill Analysis')
```

```python
# color 
tip.plot(kind = 'scatter', x = 'total_bill', y = 'tip', title = 'Restaurant Bill Analysis', color = 'purple')
```

```python
# marker
tip.plot(kind = 'scatter', x = 'total_bill', y = 'tip', title = 'Restaurant Bill Analysis', color = 'purple', marker = '+')
```

```python
# figsize
tip.plot(kind = 'scatter', x = 'total_bill', y = 'tip', title = 'Restaurant Bill Analysis', color = 'purple', marker = '+', figsize= (10, 5))
```

```python
tip['size'] = tip['size']*100
tip
```

```python
# size
tip.plot(kind = 'scatter', x = 'total_bill', y = 'tip', title = 'Restaurant Bill Analysis', color = 'red', marker = '+', figsize= (10, 5), s = 'size')
```

```python
# gender wise color
tip.plot(kind = 'scatter', x = 'total_bill', y = 'tip', title = 'Restaurant Bill Analysis', marker = '+', figsize= (10, 5), s = 'size', c = 'sex', cmap = 'Dark2')
```

#### Pie Chart

```python
df = pd.DataFrame(
    {
        'batsman':['Aaa','Bbb','Ccc','Ddd','Eee','Fff'],
        'match1':[120,90,35,45,12,10],
        'match2':[0,1,123,130,34,45],
        'match3':[50,24,145,45,10,90]
    }
)

df.head()
```

```python
# single
df['match1'].plot(kind = 'pie', labels = df['batsman'].values, autopct = '%0.1f%%')
```

```python
# multiple pie charts
df[['match1', 'match2', 'match3']].plot(kind = 'pie', subplots = True)
```

```python
df[['match1', 'match2', 'match3']].plot(kind = 'pie', subplots = True, figsize = (15, 10))
```

#### On MultiIndex DataFrame

```python
tip
```

```python
df_tip = tip.pivot_table(index = 'day', columns = 'sex', values = 'total_bill', aggfunc = 'mean')
df_tip
```

```python
df_tip.plot(kind = 'line')
```

```python
df_tip = tip.pivot_table(index = ['day', 'time'], columns = ['sex', 'smoker'], values = 'total_bill', aggfunc = 'mean')
df_tip
```

```python
df_tip.plot(kind = 'line')
```

```python
df_tip.plot(kind = 'bar')
```

```python
df_tip.plot(kind = 'pie', subplots = True, figsize = (25, 25))
```
