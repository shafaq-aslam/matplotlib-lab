# Subplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
batter = pd.read_csv('batter.csv')
batter.head()
```

```python
# A diff ways to plot graphs
plt.scatter(batter['avg'], batter['strike_rate'])
```

```python
fig, ax = plt.subplots()

ax.scatter(batter['avg'], batter['strike_rate'])
```

```python
fig, ax = plt.subplots()

ax.scatter(batter['avg'], batter['strike_rate'])
ax.set_title('Batters')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')
```

```python
fig, ax = plt.subplots(figsize=(15, 7))

ax.scatter(batter['avg'], batter['strike_rate'])
ax.set_title('Batters')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')
```

```python
fig, ax = plt.subplots(figsize=(15, 7))

ax.scatter(batter['avg'], batter['strike_rate'])
ax.set_title('Batters')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 7))

ax.scatter(batter['avg'], batter['strike_rate'], marker = '+', color='red')
ax.set_title('Batters')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')
```

```python
# why we use subplot?
# to show graphs side by side

plt.subplots(nrows=2, ncols=1)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
ax[0].scatter(batter['avg'], batter['strike_rate'], color = 'purple')
ax[1].scatter(batter['avg'], batter['runs'], color = 'green')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
ax[0].scatter(batter['avg'], batter['strike_rate'], color = 'purple')
ax[1].scatter(batter['avg'], batter['runs'], color = 'green')

ax[0].set_title('Average Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')
ax[1].set_title('Average Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharex = True)
ax[0].scatter(batter['avg'], batter['strike_rate'], color = 'green')
ax[1].scatter(batter['avg'], batter['runs'], color = 'orange')

ax[0].set_title('Average Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')
ax[1].set_title('Average Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharex = True, figsize=(10, 6))
ax[0].scatter(batter['avg'], batter['strike_rate'], color = 'blue', marker = '+')
ax[1].scatter(batter['avg'], batter['runs'], color = 'red', marker = '+')

ax[0].set_title('Average Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')
ax[1].set_title('Average Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```
