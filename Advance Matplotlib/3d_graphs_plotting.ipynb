# 3D Graphs Plotting

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 3D Scatter Plots

```python
batter = pd.read_csv('batter.csv')
batter
```

```python
fig = plt.figure()
ax = plt.subplot(projection = '3d')
```

```python
fig = plt.figure()

ax = plt.subplot(projection = '3d')
ax.scatter3D(batter['runs'], batter['avg'], batter['strike_rate'], marker='+', color = 'purple')

ax.set_title('IPL Batsman Analysis')
ax.set_xlabel('Runs')
ax.set_ylabel('Avg')
ax.set_zlabel('Strike Rate')
```

### 3D Line Plot

```python
x = [15, 25, 45, 50]
y = [8, 6, 13, 25]
z = [20, 8, 12, 15]

fig = plt.figure()
ax = plt.subplot(projection = '3d')
ax.scatter3D(x, y, z, s=[100, 100, 100, 100], marker = 'D', color = 'red')
ax.plot3D(x, y, z, color='green')
```

### 3D Surface Plots

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x, y)
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(15, 10))
ax = plt.subplot(projection='3d')
ax.plot_surface(xx, yy, z, cmap = 'inferno')
```

```python
fig = plt.figure(figsize=(15, 10))
ax = plt.subplot(projection='3d')
clr = ax.plot_surface(xx, yy, z, cmap = 'inferno')
fig.colorbar(clr)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(15, 10))
ax = plt.subplot(projection='3d')
clr = ax.plot_surface(xx, yy, z, cmap = 'spring')
fig.colorbar(clr)
```

```python
z = np.sin(xx) + np.log(yy)
fig = plt.figure(figsize=(15, 10))
ax = plt.subplot(projection='3d')
clr = ax.plot_surface(xx, yy, z, cmap = 'Dark2')
fig.colorbar(clr)
```

### Contour Plots

```python
z = xx**2 + yy**2
fig = plt.figure(figsize=(15, 10))
ax = plt.subplot(projection='3d')
clr = ax.plot_surface(xx, yy, z, cmap = 'viridis')
fig.colorbar(clr)
```

```python
fig = plt.figure(figsize=(15, 10))
ax = plt.subplot()
clr = ax.contour(xx, yy, z, cmap = 'Dark2')
fig.colorbar(clr)
```

```python
fig = plt.figure(figsize=(15, 10))
ax = plt.subplot()
clr = ax.contourf(xx, yy, z, cmap = 'Dark2')
fig.colorbar(clr)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(15, 10))
ax = plt.subplot()
clr = ax.contourf(xx, yy, z, cmap = 'spring')
fig.colorbar(clr)
```
