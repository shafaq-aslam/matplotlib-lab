# Advanced Scatter Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Colored Scatterplots

```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
# for give color to each data point
iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-virginica':1, 'Iris-versicolor':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='inferno')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='inferno')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='viridis', alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

## Plot Size

```python
plt.figure(figsize = (20, 7)) # applicable on all graphs

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='viridis', alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

## Annotation

```python
batter = pd.read_csv('batter.csv')
batter.head(5)
```

```python
sample_df = batter.head(100).sample(25, random_state=5)
sample_df
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])
```

```python
# to place text on a graph to identify the points
x = [2, 5, 6, 9]
y = [3, 6, 7, 10]

plt.scatter(x,y)
plt.text(2,3, 'point 1')
plt.text(5,6, 'point 2')
plt.text(6,7, 'point 3')
plt.text(9,10, 'point 4')
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
plt.figure(figsize=(15, 7))
plt.scatter(sample_df['avg'], sample_df['strike_rate'])

for i in range(len(sample_df)):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
plt.figure(figsize=(25, 12))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

for i in range(len(sample_df)):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

## Horizontal and Vertical Lines

```python
plt.figure(figsize=(25, 12))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

plt.axhline(125, color='red')
plt.axhline(140, color='green')
plt.axvline(30, color='red')

for i in range(len(sample_df)):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```
