# Heatmap

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ipl_ball_data = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
ipl_ball_data.head()
```

```python
ipl_ball_data['ballnumber'].unique()
```

```python
temp_ipl_bdata = ipl_ball_data[(ipl_ball_data['ballnumber'].isin([1, 2, 3, 4, 5, 6])) & (ipl_ball_data['batsman_run']==6)]
temp_ipl_bdata
```

```python
temp_ipl_bdata.pivot_table(index = 'overs', columns = 'ballnumber', values = 'batsman_run', aggfunc = 'count')
```

```python
grid = temp_ipl_bdata.pivot_table(index = 'overs', columns = 'ballnumber', values = 'batsman_run', aggfunc = 'count')
```

```python
plt.figure(figsize = (15, 30))
plt.imshow(grid)
plt.colorbar()
```

```python
plt.figure(figsize = (15, 30))
plt.imshow(grid)
plt.yticks(ipl_ball_data['overs'].unique(), list(range(1, 21)))
plt.colorbar()
```

```python
plt.figure(figsize = (15, 30))
plt.imshow(grid)
plt.yticks(ipl_ball_data['overs'].unique(), list(range(1, 21)))
plt.xticks(np.arange(0, 6), list(range(1, 7)))
plt.colorbar()
```
