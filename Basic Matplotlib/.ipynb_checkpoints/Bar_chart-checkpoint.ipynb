```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# simple bar chart
children = [10, 15, 12, 23, 18, 9, 20]
colors = ['red', 'blue', 'green', 'black', 'yellow', 'orange', 'purple']

plt.bar(colors, children, color = 'purple')
```

```python
# horizontal bar chart
plt.barh(colors, children, color = 'green')
```

```python
# bar chart using data
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'], df['2015'], width=0.3)
```

```python
# Multiple bar chart
plt.bar(np.arange(df.shape[0]) - .2, df['2015'], width = .2)
plt.bar(np.arange(df.shape[0]), df['2016'], width = .2)
plt.bar(np.arange(df.shape[0]) + .2, df['2017'], width = .2)
```

```python
# xticks
plt.bar(np.arange(df.shape[0]) - .2, df['2015'], width = .2)
plt.bar(np.arange(df.shape[0]), df['2016'], width = .2)
plt.bar(np.arange(df.shape[0]) + .2, df['2017'], width = .2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])
```

```python
# color and label
plt.bar(np.arange(df.shape[0]) - .2, df['2015'], width = .2, color = 'yellow')
plt.bar(np.arange(df.shape[0]), df['2016'], width = .2, color = 'red')
plt.bar(np.arange(df.shape[0]) + .2, df['2017'], width = .2, color = 'green')

plt.xticks(np.arange(df.shape[0]), df['batsman'])
```

```python
# problem ---> label overlapping
children = [10, 15, 12, 23, 18, 9, 20]
colors = ['red is reedddd but Red is also red', 'blue is bluuee but Blue is also blue', 'green is greeenn but Green is also green', 'black is blaaacckk but Black is also black', 'yellow is yellowww but Yellow is also yellow', 'orange is orangggeee but Orange is also orange', 'purple is purplleee but Purple is also purple']

plt.bar(colors, children, color = 'Orange')
```

```python
# solution
children = [10, 15, 12, 23, 18, 9, 20]
colors = ['red is reedddd but Red is also red', 'blue is bluuee but Blue is also blue', 'green is greeenn but Green is also green', 'black is blaaacckk but Black is also black', 'yellow is yellowww but Yellow is also yellow', 'orange is orangggeee but Orange is also orange', 'purple is purplleee but Purple is also purple']

plt.bar(colors, children, color = 'violet')
plt.xticks(rotation = 'vertical')
```

### Stacked bar chart

```python
plt.bar(df['batsman'], df['2015'], label = '2015')
plt.bar(df['batsman'], df['2016'], bottom = df['2015'], label = '2016')
plt.bar(df['batsman'], df['2017'], bottom = (df['2016'] + df['2015']), label = '2017')

plt.legend()
```
