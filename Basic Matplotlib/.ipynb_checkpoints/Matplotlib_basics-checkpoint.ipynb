## 2D Line Plot

When we use 2D plot?

<ul>
    <li>Bivariate Analysis</li>
    <li>Categorical Analysis --> Numerical and numerical --> Categorical and numerical --> Numerical</li>
    <li>Use case --> Time series data</li>
</ul>

```python
# plotting a simple function
price = [45000, 58000, 65000, 59000, 56000, 78000, 54000, 48000, 42000, 46000]
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]

plt.plot(year, price)
```

```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# label title
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'], batsman['V Kohli'], color ='green')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'red')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color ='#07DBE5')
plt.plot(batsman['index'], batsman['RG Sharma'], color = '#F00EAF')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# line style
plt.plot(batsman['index'], batsman['V Kohli'], color ='#07DBE5', linestyle = 'dashed')
plt.plot(batsman['index'], batsman['RG Sharma'], color = '#F00EAF', linestyle = 'dotted')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

# line styles: dashed, dotted, dashdot, etc.
```

```python
# line width
plt.plot(batsman['index'], batsman['V Kohli'], color ='orange', linestyle = 'solid', linewidth = 3)
plt.plot(batsman['index'], batsman['RG Sharma'], color = '#80045B', linestyle = 'dotted', linewidth = 2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# marker
plt.plot(batsman['index'], batsman['V Kohli'], color ='orange', linestyle = 'solid', linewidth = 3, marker = 'D')
plt.plot(batsman['index'], batsman['RG Sharma'], color = '#80045B', linestyle = 'dotted', linewidth = 2, marker = 'o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# marker size
plt.plot(batsman['index'], batsman['V Kohli'], color ='orange', linestyle = 'solid', marker = 'D', markersize = 10)
plt.plot(batsman['index'], batsman['RG Sharma'], color = '#80045B', linestyle = 'dotted', marker = 'o', markersize = 10)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend --> location
plt.plot(batsman['index'], batsman['V Kohli'], color ='orange', linestyle = 'solid', marker = 'D', markersize = 5, label = 'V Kohli')
plt.plot(batsman['index'], batsman['RG Sharma'], color = '#80045B', linestyle = 'dotted', marker = 'o', markersize = 5, label = 'RG Sharma')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color ='orange', linestyle = 'solid', marker = 'D', markersize = 5, label = 'V Kohli')
plt.plot(batsman['index'], batsman['RG Sharma'], color = '#80045B', linestyle = 'dotted', marker = 'o', markersize = 5, label = 'RG Sharma')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc = 'upper center')
```
