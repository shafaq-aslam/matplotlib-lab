When we use Pie Chart?

<ul>
    <li>Univariate/Bivariate Analysis</li>
    <li>Categorical vs numerical</li>
    <li>Use case - To find contibution on a standard scale</li>
</ul>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# simple data
data = [25, 65, 45, 85, 95, 10]
plt.pie(data)
plt.show()
```

```python
data = [25, 65, 45, 80, 95, 10]
subject = ['Eng', 'Urdu', 'Hist', 'Math', 'Comp', 'Geo']
plt.pie(data, labels = subject)
plt.show()
```

```python
# dataset
df = pd.read_csv('gayle-175.csv')
plt.pie(df['batsman_runs'], labels = df['batsman'])
plt.show()
```

```python
df = pd.read_csv('gayle-175.csv')
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%')
plt.show()
```

```python
# percentage and colors
df = pd.read_csv('gayle-175.csv')
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%', colors=['cyan', 'green', 'yellow', 'orange', 'red', 'purple'])
plt.show()
```

```python
# explode
df = pd.read_csv('gayle-175.csv')
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%', explode=[0.1,0,0,0.2,0,0])
plt.show()
```

```python
# shadow
df = pd.read_csv('gayle-175.csv')
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%', explode=[0.1,0,0,0.2,0,0], shadow=True)
plt.show()
```
