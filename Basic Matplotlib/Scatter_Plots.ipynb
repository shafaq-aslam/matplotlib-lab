```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plt.scatter simple function
arrx = np.linspace(-10, 10, 50)
arrx
```

```python
arry = 10*arrx + 3 + np.random.randint(0, 300, 50)
arry
```

```python
plt.scatter(arrx, arry)
```

```python
# plt.scatter on pandas data
match_df = pd.read_csv('batter.csv')
df = match_df.head(50)
df
```

```python
# Labeling
plt.scatter(df['avg'], df['strike_rate'])
plt.title('Avg and Strike_Rate Analysis of Top 50 Batsmen')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

```python
# Marker 
plt.scatter(df['avg'], df['strike_rate'], color = 'green', marker = '+')
plt.title('Avg and Strike_Rate Analysis of Top 50 Batsmen')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

```python
# Size- Plot
tips = sns.load_dataset('tips')

plt.scatter(tips['total_'])
```
